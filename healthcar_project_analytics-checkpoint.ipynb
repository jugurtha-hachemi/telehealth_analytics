```python
# ----------------------------
# Synthetic Healthcare Project Dataset Generator
# Moderate Telehealth Adoption
# ----------------------------
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()
np.random.seed(42)
# ------ save in datasets
os.makedirs("C:/analytics/data", exist_ok=True)
# ----------------------------
# 1Ô∏è‚É£ Patients Dataset
# ----------------------------
num_patients = 10000
patients = pd.DataFrame({
    "patient_id": np.arange(100000, 100000 + num_patients),
    "age": np.random.randint(18, 90, size=num_patients),
    "gender": np.random.choice(["M", "F"], size=num_patients),
    "region": np.random.choice(["MN","TX","CA","FL","NY","AZ","WA","IL","NV","GA"], size=num_patients),
    "chronic_flag": np.random.choice([0,1], size=num_patients, p=[0.6,0.4]),
    "insurance_plan": np.random.choice(["HMO_A","PPO_B","EPO_C"], size=num_patients)
})
patients.to_csv("C:/analytics/data/patients.csv", index=False)

# ----------------------------
# 2Ô∏è‚É£ Providers Dataset
# ----------------------------
num_providers = 50
providers = pd.DataFrame({
    "provider_id": np.arange(2000, 2000+num_providers),
    "specialty": np.random.choice(["Primary Care","Endocrinology","Cardiology","Pulmonology","Family Medicine","Emergency Medicine"], size=num_providers),
    "region": np.random.choice(["MN","TX","CA","FL","NY","AZ","WA","IL","NV","GA"], size=num_providers),
    "network_flag": np.random.choice([0,1], size=num_providers, p=[0.1,0.9])
})
providers.to_csv("C:/analytics/data/providers.csv", index=False)

# ----------------------------
# 3Ô∏è‚É£ Telehealth Visits (Moderate Adoption ~45%)
# ----------------------------
telehealth_patients = patients.sample(frac=0.45, random_state=42)["patient_id"]
num_telehealth_visits = 25000
telehealth_visits = pd.DataFrame({
    "visit_id": np.arange(500000, 500000 + num_telehealth_visits),
    "patient_id": np.random.choice(telehealth_patients, size=num_telehealth_visits),
    "visit_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, size=num_telehealth_visits), unit='D'),
    "visit_type": np.random.choice(["Video","Phone"], size=num_telehealth_visits, p=[0.7,0.3]),
    "provider_id": np.random.choice(providers["provider_id"], size=num_telehealth_visits),
    "visit_cost": np.round(np.random.uniform(50, 200, size=num_telehealth_visits),2)
})
telehealth_visits.to_csv("C:/analytics/data/telehealth_visits.csv", index=False)

# ----------------------------
# 4Ô∏è‚É£ ER Visits Dataset
# ----------------------------
num_er = 15000
er_visits = pd.DataFrame({
    "er_id": np.arange(900000, 900000 + num_er),
    "patient_id": np.random.choice(patients["patient_id"], size=num_er),
    "er_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, size=num_er), unit='D'),
    "total_cost": np.round(np.random.uniform(500, 10000, size=num_er),2),
    "admission_flag": np.random.choice([0,1], size=num_er, p=[0.7,0.3]),
    "claim_submit_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, size=num_er), unit='D')
})
er_visits.to_csv("C:/analytics/data/er_visits.csv", index=False)

# ----------------------------
# 5Ô∏è‚É£ Pharmacy Claims
# ----------------------------
num_rx = 20000
drugs = ["Lisinopril","Metformin","Albuterol","Furosemide","Atorvastatin","Insulin"]
pharmacy_claims = pd.DataFrame({
    "rx_id": np.arange(300000, 300000 + num_rx),
    "patient_id": np.random.choice(patients["patient_id"], size=num_rx),
    "drug_name": np.random.choice(drugs, size=num_rx),
    "drug_tier": np.random.randint(1,4, size=num_rx),
    "generic_flag": np.random.choice([0,1], size=num_rx, p=[0.3,0.7]),
    "fill_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, size=num_rx), unit='D'),
    "drug_cost": np.round(np.random.uniform(5,150,size=num_rx),2)
})
pharmacy_claims.to_csv("C:/analytics/data/pharmacy_claims.csv", index=False)

# ----------------------------
# 6Ô∏è‚É£ Claims Dataset
# ----------------------------
num_claims = 50000
claims = pd.DataFrame({
    "claim_id": np.arange(800000, 800000+num_claims),
    "patient_id": np.random.choice(patients["patient_id"], size=num_claims),
    "provider_id": np.random.choice(providers["provider_id"], size=num_claims),
    "claim_amount": np.round(np.random.uniform(100,10000,size=num_claims),2),
    "submission_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, size=num_claims), unit='D'),
    "adjudication_date": pd.to_datetime("2023-01-02") + pd.to_timedelta(np.random.randint(1,30,size=num_claims), unit='D'),
    "denial_flag": np.random.choice([0,1], size=num_claims, p=[0.8,0.2]),
    "denial_reason": np.random.choice([None,"Technical","Clinical","Administrative"], size=num_claims)
})
claims.to_csv("C:/analytics/data/claims.csv", index=False)

# ----------------------------
# 7Ô∏è‚É£ Provider Claim Summary (Aggregated CMS-style)
# ----------------------------
provider_claim_summary = claims.groupby("provider_id").agg(
    total_claims=("claim_id","count"),
    total_claim_amount=("claim_amount","sum"),
    total_denials=("denial_flag","sum")
).reset_index()
provider_claim_summary = provider_claim_summary.merge(providers[["provider_id","specialty","region","network_flag"]], on="provider_id")
provider_claim_summary.to_csv("C:/analytics/data/provider_claim_summary.csv", index=False)

print("‚úÖ All 7 CSVs generated for moderate telehealth adoption!")
```

```python
# Load datasets
patients = pd.read_csv("patients.csv")
telehealth = pd.read_csv("telehealth_visits.csv")
```

```python
# Count telehealth visits per patient
telehealth_counts = telehealth.groupby("patient_id").size().reset_index(name="num_visits")
```

```python
# Merge with patient demographics
telehealth_summary = patients.merge(telehealth_counts, on="patient_id", how="left")
```

```python
# Fill patients with 0 visits
telehealth_summary["num_visits"] = telehealth_summary["num_visits"].fillna(0).astype(int)
```

```python
# Preview
telehealth_summary.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create age groups
bins = [18, 30, 40, 50, 60, 70, 80, 90]
labels = ["18-29","30-39","40-49","50-59","60-69","70-79","80-89"]
telehealth_summary["age_group"] = pd.cut(telehealth_summary["age"], bins=bins, labels=labels, right=False)

# Aggregate adoption metrics
adoption_stats = telehealth_summary.copy()
adoption_stats["is_user"] = adoption_stats["num_visits"] > 0

# 1Ô∏è‚É£ Telehealth adoption by age group
age_adoption = adoption_stats.groupby("age_group")["is_user"].mean().reset_index()
age_adoption["adoption_percent"] = age_adoption["is_user"] * 100

# 2Ô∏è‚É£ Telehealth adoption by gender
gender_adoption = adoption_stats.groupby("gender")["is_user"].mean().reset_index()
gender_adoption["adoption_percent"] = gender_adoption["is_user"] * 100

# 3Ô∏è‚É£ Telehealth adoption by chronic condition
chronic_adoption = adoption_stats.groupby("chronic_flag")["is_user"].mean().reset_index()
chronic_adoption["adoption_percent"] = chronic_adoption["is_user"] * 100
chronic_adoption["chronic_flag"] = chronic_adoption["chronic_flag"].map({0:"No Chronic",1:"Chronic"})

# -------------------
# Visualizations
# -------------------
plt.figure(figsize=(16,4))

# Age group
plt.subplot(1,3,1)
sns.barplot(x="age_group", y="adoption_percent", data=age_adoption, palette="Blues_d")
plt.title("Telehealth Adoption by Age Group")
plt.ylabel("Adoption %")
plt.xlabel("Age Group")

# Gender
plt.subplot(1,3,2)
sns.barplot(x="gender", y="adoption_percent", data=gender_adoption, palette="Greens_d")
plt.title("Telehealth Adoption by Gender")
plt.ylabel("Adoption %")
plt.xlabel("Gender")

# Chronic condition
plt.subplot(1,3,3)
sns.barplot(x="chronic_flag", y="adoption_percent", data=chronic_adoption, palette="Reds_d")
plt.title("Telehealth Adoption by Chronic Condition")
plt.ylabel("Adoption %")
plt.xlabel("Chronic Condition")

plt.tight_layout()
plt.show()
```

```python
# Load ER visits dataset
er_visits = pd.read_csv("er_visits.csv")

# Identify telehealth users
telehealth_summary["is_user"] = telehealth_summary["num_visits"] > 0

# Merge ER visits with patient data
er_merged = er_visits.merge(telehealth_summary[["patient_id","chronic_flag","is_user"]], on="patient_id", how="left")

# Count ER visits per patient
er_counts = er_merged.groupby(["patient_id","chronic_flag","is_user"]).size().reset_index(name="er_visits")

# For patients with chronic conditions only
chronic_patients_er = er_counts[er_counts["chronic_flag"]==1]

# Average ER visits by telehealth usage
er_summary = chronic_patients_er.groupby("is_user")["er_visits"].mean().reset_index()
er_summary["telehealth_status"] = er_summary["is_user"].map({True:"Telehealth User", False:"Non-user"})

print(er_summary)
```

```python
sns.barplot(x="telehealth_status", y="er_visits", data=er_summary, palette="coolwarm")
plt.title("Average ER Visits for Chronic Patients by Telehealth Usage")
plt.ylabel("Average ER Visits")
plt.show()
```

```python
# Load telehealth and ER datasets
telehealth = pd.read_csv("telehealth_visits.csv")
er_visits = pd.read_csv("er_visits.csv")

# Average cost per telehealth visit
avg_telehealth_cost = telehealth["visit_cost"].mean()

# Average cost per ER visit
avg_er_cost = er_visits["total_cost"].mean()

# Potential savings per visit
savings_per_visit = avg_er_cost - avg_telehealth_cost

print(f"Average Telehealth Visit Cost: ${avg_telehealth_cost:.2f}")
print(f"Average ER Visit Cost: ${avg_er_cost:.2f}")
print(f"Average Savings per Visit if Telehealth Used: ${savings_per_visit:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting
cost_df = pd.DataFrame({
    "Visit Type": ["Telehealth","ER"],
    "Average Cost": [avg_telehealth_cost, avg_er_cost]
})

# Bar plot
sns.barplot(x="Visit Type", y="Average Cost", data=cost_df, palette="Set2")
plt.title("Average Cost Comparison: Telehealth vs ER")
plt.ylabel("Cost ($)")
plt.show()
```
